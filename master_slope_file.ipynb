```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Define the lag value
lag = 14  # Example lag value, adjust as needed

# Assuming onset_df, df_date, and other variables are defined earlier

# Ensure 'Onset_Date' is in datetime format
onset_df['Onset_Date'] = pd.to_datetime(onset_df['Onset_Date'])

# Remove duplicate rows based on the 'Cyclone' column
onset_df = onset_df.drop_duplicates(subset='Cyclone', keep='first')

# Filter unique cyclone names based on onset dates after 2011
valid_cyclone_names = onset_df[onset_df['Onset_Date'].dt.year >= 2012]['Cyclone'].unique()

# Initialize an empty DataFrame to store the results
final_df = pd.DataFrame(columns=['Cyclone', 'Settle_PCOD', 'hist_pre_average', 'hist_post_average', 'cyclone_pre_average', 'cyclone_post_average'])

# Initialize dictionaries to store mean values for cyclone-related columns
cyclone_pre_avg_dict = {}
cyclone_post_avg_dict = {}

# Iterate over each valid cyclone name
for cyclone_name in valid_cyclone_names:
    # Initialize dictionaries to store mean values for this cyclone
    hist_pre_avg_dict = {}
    hist_post_avg_dict = {}
    
    # Iterate over each settle_pcod value
    for settle_pcod in df_date.index:
        result_dict = {'Settle_PCOD': settle_pcod, 'Cyclone': cyclone_name}
        
        # Check if cyclone onset information exists
        if cyclone_name in onset_df['Cyclone'].values:
            cyclone_onset_date = onset_df[onset_df['Cyclone'] == cyclone_name]['Onset_Date'].iloc[0]
            cyclone_onset_year = cyclone_onset_date.year

            # Pre-cyclone period calculation
            pre_start_date = pd.Timestamp(year=cyclone_onset_year, month=cyclone_onset_date.month, day=cyclone_onset_date.day) - pd.Timedelta(days=lag)
            pre_end_date = pd.Timestamp(year=cyclone_onset_year, month=cyclone_onset_date.month, day=cyclone_onset_date.day)
            pre_date_mask_range = pd.date_range(start=pre_start_date, end=pre_end_date)
            pre_filtered_dates = df_date.columns[df_date.columns.isin(pre_date_mask_range)]
            if not pre_filtered_dates.empty:
                pre_mean_values = df_date.loc[settle_pcod, pre_filtered_dates].mean()
                result_dict['hist_pre_average'] = pre_mean_values
                hist_pre_avg_dict[settle_pcod] = pre_mean_values
            else:
                print(f"No data available for the specified pre date range in {cyclone_onset_year}.")
        
            # Post-cyclone period calculation
            post_start_date = pd.Timestamp(year=cyclone_onset_year, month=cyclone_onset_date.month, day=cyclone_onset_date.day)
            post_end_date = post_start_date + pd.Timedelta(days=lag)
            post_date_mask_range = pd.date_range(start=post_start_date, end=post_end_date)
            post_filtered_dates = df_date.columns[df_date.columns.isin(post_date_mask_range)]
            if not post_filtered_dates.empty:
                post_mean_values = df_date.loc[settle_pcod, post_filtered_dates].mean()
                result_dict['hist_post_average'] = post_mean_values
                hist_post_avg_dict[settle_pcod] = post_mean_values
            else:
                print(f"No data available for the specified post date range in {cyclone_onset_year}.")

        else:
            print(f"Cyclone '{cyclone_name}' not found in the dataset.")

        # Append the result dictionary to the final DataFrame
        final_df = final_df.append(result_dict, ignore_index=True)

    # Calculate mean values for cyclone-related columns for this cyclone
    if hist_pre_avg_dict:
        cyclone_pre_avg_dict[cyclone_name] = sum(hist_pre_avg_dict.values()) / len(hist_pre_avg_dict)
    else:
        cyclone_pre_avg_dict[cyclone_name] = float('nan')
    
    if hist_post_avg_dict:
        cyclone_post_avg_dict[cyclone_name] = sum(hist_post_avg_dict.values()) / len(hist_post_avg_dict)
    else:
        cyclone_post_avg_dict[cyclone_name] = float('nan')

# Fill cyclone-related columns with mean values for each cyclone
final_df['cyclone_pre_average'] = final_df['Cyclone'].map(cyclone_pre_avg_dict)
final_df['cyclone_post_average'] = final_df['Cyclone'].map(cyclone_post_avg_dict)

# Reorder columns
final_df = final_df[['Cyclone', 'Settle_PCOD', 'hist_pre_average', 'hist_post_average', 'cyclone_pre_average', 'cyclone_post_average']]

# Define a function to calculate slope
def calculate_slope(average_value, cyclone_onset_date, lag, df_date, settle_pcod):
    cyclone_onset_year = cyclone_onset_date.year
    
    # Calculate start and end dates based on lag
    pre_start_date = pd.Timestamp(year=cyclone_onset_year, month=cyclone_onset_date.month, day=cyclone_onset_date.day) - pd.Timedelta(days=lag)
    pre_end_date = pd.Timestamp(year=cyclone_onset_year, month=cyclone_onset_date.month, day=cyclone_onset_date.day)
    pre_date_range = pd.date_range(start=pre_start_date, end=pre_end_date)

    # Check if the average value is NaN
    if pd.isna(average_value):
        return float('nan')
    
    # Filter the pre_date_range to only include existing dates in df_date index
    pre_date_range = pre_date_range[pre_date_range.isin(df_date.columns)]
    
    # Check if pre_date_range contains any dates
    if pre_date_range.empty:
        return float('nan')
    
    # Check if pre_date_range contains NaN values
    if df_date.loc[settle_pcod, pre_date_range].isnull().any():
        return float('nan')
    
    # Prepare X and y data for linear regression
    X = np.arange(len(pre_date_range)).reshape(-1, 1)
    y = df_date.loc[settle_pcod, pre_date_range].values.reshape(-1, 1)

    # Fit linear regression model
    model = LinearRegression().fit(X, y)
    
    # Return the slope of the regression line
    return model.coef_[0][0]

# Apply the function to calculate slope and add it as a column
final_df['cyclone_pre_average_slope'] = final_df.apply(lambda row: calculate_slope(row['cyclone_pre_average'], onset_df.loc[onset_df['Cyclone'] == row['Cyclone'], 'Onset_Date'].iloc[0], lag, df_date, row['Settle_PCOD']), axis=1)
final_df['cyclone_post_average_slope'] = final_df.apply(lambda row: calculate_slope(row['cyclone_post_average'], onset_df.loc[onset_df['Cyclone'] == row['Cyclone'], 'Onset_Date'].iloc[0], lag, df_date, row['Settle_PCOD']), axis=1)

# Add the 'Onset_Date' column to the final_df DataFrame
final_df['Onset_Date'] = final_df['Cyclone'].map(onset_df.set_index('Cyclone')['Onset_Date'])

# Write the DataFrame to a CSV file
final_df.to_csv(belmontdir + 'slope_all_data.csv', index=False)

print("CSV file 'slope_all_data.csv' has been created.")
```
